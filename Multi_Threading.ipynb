# Multithreading In Python

image.png

Multithreading is a concurrent execution technique which allows a python program to perform multiple tasks concurrently. In simple words, we know that our computers have multiple cores, some have four (quad-core processor) while some have eight core (octa-core processor). The programs we develop and run are running on these cores, one program per core at one time. Now, if we want to run multiple programs on one single core, keeping rest of the cores free for other work, we can do that using the concept of threading.

In multithreading, a program consists of multiple threads, each of which executes independently but shares the same resources like memory space. Python's threading module is used to create and manage threads.

But, How to use multithreading in python? A simplifies overview of this is explained in the following cells of code.

```python
# import the threading module
import threading
import time
```

```python
# defining a function for the tread
def my_func(x):
  # the logic or the body of the function
  print('this is thread ', x)
```

```python
# creating threads
thread01 = [threading.Thread(target=my_func, args = (i ,)) for i in [1,2,3,4]]

#start thread
for t in thread01:
  t.start()
```

The above example of threading in python is a simple use of the concept of threading for running a program multiple times, with a different argument each time and also in one single core.

Now, we can also use the concept of threading in many other tasks, one of which is to fetch data from urls into different text files. We want to do this using threading, which means we will be writing one piece of code which will be fetching data from all three urls, into three different text files one by one.

For this, we first need the urls. The ones I am using here are:

1. https://raw.githubusercontent.com/itsfoss/text-files/master/agatha.txt
2. https://raw.githubusercontent.com/itsfoss/text-files/master/sherlock.txt
3. https://raw.githubusercontent.com/itsfoss/text-files/master/sample_log_file.txt

Also, as we need to fetch the data from online, urls, we need the python library 'urllib.request' to access these urls and allow us to perform the data extraction part.

The steps we will be following to perform this task are:

1. import urllib and create a function to fetch the data.


```python
import urllib.request

def file_download(url , filename):
    urllib.request.urlretrieve(url , filename)
```

2. create a list of all the links we need to fetch the data from.


```python
url_list = ['https://raw.githubusercontent.com/itsfoss/text-files/master/agatha.txt' , 'https://raw.githubusercontent.com/itsfoss/text-files/master/sherlock.txt' ,'https://raw.githubusercontent.com/itsfoss/text-files/master/sample_log_file.txt' ]
```

```python
url_list
```

3. create a list of the text files


```python
data_file_list = ['data1.txt', 'data2.txt','data3.txt']
```

4. create the thread using the function, url list, and text files list.


```python
thread2 = [threading.Thread(target=file_download , args=(url_list[i],data_file_list[i])) for i in range(len(url_list))]
```

5. run the thread and you have the files ready all at once.

```python
for t in thread2:
    t.start()
```

What else can we do using threading?
We can also use 'time' in our threads to create a unique processing based on time. Following example shows us how we can do that.

Let's try to write a simple function which just prints a text, and try to incorporate time into it to control the processing of the function using thread.

For this, we need to import time first and then define our function to then create the thread. Let's do it and see what happens.

```python
import time
```

```python
def time_test(x) :
    for i in range(5) :
        print(" This is the thread for the value x %d and for the value of i %d " %(x,i))
        # time.sleep(1)
```

```python
thread3 = [threading.Thread(target=time_test , args=(i,)) for i in [100 , 10,20,5]]
```

```python
for t in thread3:
    t.start()
```

```python
def test2(x) :
    for i in range(5) :
        print(" test1 print the value of x %d and print the value of i %d " %(x,i))
        time.sleep(0.05)
```

```python
thread4 = [threading.Thread(target=test2 , args=(i,)) for i in [100 , 10,20,5]]
```

```python
for t in thread4:
    t.start()
```

In this example, we saw that the one line code 'time.sleep()' plays an important role in how the thred is processing the code. When we are not including the sleep time, the thread process the code for all i values and one x value. On the other hand, when we include the sleep time, the thread process one i value and all x values.

This difference of processing the same code in two different formats is very crucial observation and can be used in various real life situations such as task scheduling, asynchronous I/O operations, concurrent API requests and more.

### Best Practices for Multithreading

1. Avoid Global State: Minimize the use of global variables and shared state between threads, as it can lead to race conditions and data corruption.

2. Thread Safety: We should ensure that data shared between threads is accessed in a thread-safe manner using locks or other sunchronization mechanisms.

3. Resource Cleanup: Properly release resources i.e, file, sockets or more, when threads are done using them. Use 'try-finally' blocks to ensure cleanup.

4. Use Thread Pools: For better resource management, consider using thread pools provided by libraries like 'concurrent.futures'.

5. Profile and Monitor: Profile your multithreaded code to identify performance bottlenecks. Monitor thread activities to detect any deadlocks or other issues.

6. GIL Limitation: We must keep inn mind that Python has a Global Interpreter Lock (GIL), which prevents true parallel execution of threads. For this, we can use multiprocessing for CPU bound tasks if and when necessary.

Now, we can use 'mutex' short for 'mutual exclusion' in python to protect the shared resources from being accessed by multiple threads at the same time. This helps in preventing race conditions and data corruption by providing exclusive access to the protected resources. Threads should/must acquire the mutex lock before accessing the resource and release it when the task is done, allowing other threads to acquire it.

Multithreading can be a powerful technique to improve the concurrency and performance of your Python programs, but it should be used carefully to avoid threading-related issues. Understanding thread safety and following best practices is crucial when working with multithreaded applications.
