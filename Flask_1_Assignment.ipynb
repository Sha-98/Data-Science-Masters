## **Q1. What is Flask Framework? What are the advantages of Flask Framework?**

**Flask Framework -** Flask is a lightweight and micro web framework written in Python. It is designed to be simple, easy to use, and extensible. Flask provides the essentials for building web applications without imposing a specific project structure or dependencies. It follows the WSGI (Web Server Gateway Interface) standard, making it compatible with various web servers.

Key features of Flask include:

* **Routing:** Flask allows developers to define URL patterns and associate them with Python functions, making it easy to create routes for different parts of a web application.

* **Template Engine:** Flask includes Jinja2, a powerful and expressive template engine, for rendering HTML templates.

* **Web Server:** While Flask comes with a built-in development server, it can also be deployed with other WSGI-compliant servers, such as Gunicorn or uWSGI, for production environments.

* **Extensions:** Flask has a modular design, and developers can choose from a variety of extensions to add features like authentication, database integration, and more.

* **Minimalistic:** Flask follows the "micro" philosophy, providing only the essentials for web development. This allows developers to choose and integrate additional components based on their specific needs.


**Advantages of Flask Framework:**

**1. Simplicity:** Flask is easy to learn and use, making it a great choice for beginners and projects with straightforward requirements.

**2. Flexibility:** Flask does not impose a specific project structure, allowing developers to choose their own architecture and libraries based on the project's needs.

**3. Lightweight:** Being a micro-framework, Flask has a small codebase and minimal dependencies, resulting in faster startup times and lower overhead.

**4. Extensibility:** Flask's modular design allows developers to add functionality through extensions, giving them the flexibility to choose the components they need.

**5. Jinja2 Templating:** Flask uses the Jinja2 templating engine, which provides powerful and flexible template rendering for dynamic content.

**6. RESTful Support:** Flask supports the development of RESTful APIs, making it versatile for both traditional web applications and API development.

**7. Active Development:** Flask is actively maintained and updated, ensuring that it keeps up with best practices and security standards.

**8. Compatibility:** Flask can be integrated with other Python libraries and tools, allowing developers to leverage a wide range of functionalities.




## **Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.**


image.png

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)
```

**The Flask web server will start running and you can access the "Hello World!!" message at http://localhost:5000/.**

The code used to create this flask app is explained below.

1. The from flask import Flask statement imports the Flask class, which is used to create the web server.

> The app = Flask(__name__) line creates a Flask web server object.

2. The @app.route('/') decorator above the hello_world() function defines that when the web server receives a request for the root URL ("/"), it should call the hello_world() function.

3. The def hello_world(): line starts the definition of the hello_world() function.

4. The return 'Hello World!!' line is the body of the hello_world() function. It tells the function to return the string "Hello World!!".

5. The if __name__='__main__': line checks if the current Python module is the main module of the program. If it is, the Flask web server will be started.

6. The app.run(debug=True) line starts the Flask web server with debugging mode enabled. This means that if any errors occur, the Flask web server will provide detailed error messages.

## **Q3. What is App routing in Flask? Why do we use app routes?**

In Flask, app routing refers to the mechanism of associating different URLs (Uniform Resource Locators) with specific Python functions, known as view functions. Routing defines how the application responds to different HTTP requests for specific paths or patterns in the URL.

In Flask, the **@app.route()** decorator is used to define routes. When a user accesses a specific URL, Flask maps that URL to the corresponding decorated view function, allowing the application to generate a response.

## **Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.**

In Flask, the **url_for()** function is used for URL building. It generates a URL for the specified endpoint (function) and any arguments provided. This function is particularly useful because it allows you to change the URL structure without having to update URLs manually throughout your application.

An example demonstrating the use of the url_for() function in Flask is given below:

```python
from flask import Flask, url_for

# Create a Flask application
app = Flask(__name__)

# Define a route for the "/welcome" URL
@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

# Define a route for the "/details" URL with a dynamic parameter
@app.route('/details/<username>')
def user_details(username):
    return f'Details for user: {username}'

# Define a route for the "/" URL
@app.route('/')
def home():
    # Using url_for() to generate URLs for the welcome and user_details endpoints
    welcome_url = url_for('welcome')
    details_url = url_for('user_details', username='john_doe')

    # Display the generated URLs
    return f'''
    Welcome URL: {welcome_url}<br>
    User Details URL: {details_url}
    '''

# Run the application if executed directly
if __name__ == '__main__':
    app.run(debug=True)
```
