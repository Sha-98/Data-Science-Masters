# API - All about HTTP Methods

### **What is an HTTP method?**
HTTP (Hypertext transfer Protocol) methods, or also HTTP verbs as said, define the actions that can be performed on resources identified by a URL.

Each HTTP method has a specific purpose and is used in different scenarios. This article talks about different, most used HTTP methods and their purposes in API.

image.png

**Some commonly used HTTP methods are listed and discusses below**

## **01. GET:**

image.png



* **Purpose: "Retrieve data from the server."**
Which means that when we use the GET method, we are like someone going to the library and asking the librarian for specific book. The librarian gives us the book and we leave with it.

* **Idempotent: Yes (repeated requests have the same effect as a single request).** Now, here the term, 'indempotent' is like saying that if we go back to the librarian and ask for the same book, we'll get the same information. It is like asking for the latest news online. If we refresh the page, we'll get the latest news again. The action of refreshing is 'indempotent'.

* **Example: Retrieving the content of a web page.** As an example, in practical terms, when we type a web address into our browser and hit 'Enter', we're actually sending a GET request to the server hosting that website. The server then responds by sending us the web page content - all the we see.  


## **02. POST:**



image.png

* **Purpose: Submit data to be processed to a specified resource.** Let's think of POST like a sending package. When we POST something, we're giving new information to the server and asking it to do something with that information. It is like sending a package to a friend; we're delivering new content, the package to our friend's loaction, the server for us and expecting some specific action to be taken.

* **Idempotent: No (repeated requests may have different effects).** Unline GET, POST request or method is not indempotent. Which means, in POST, if we send same request to the server twice, the server might react differently each time, as it might create a new user account for new request, no matter if it is the same request.

* **Example: Submitting a form with user data.** A form we submit on goolge is a good example of the POST method, as here we are making the POST request by sending the data to the sever and asking it to do something with it (it can be any task, as simple as storing also). Everytime we submit the same google form, even if we are adding same info, google considers it as a new user and saves it that way only. This is the POST request for us.

## **03. PUT:**




image.png

* **Purpose: Update a resource or create a new resource if it does not exist.** The PUT method can be thought of like updating a book. If the book already exists, we're revisiting it. If it doesn't exist, we're creating a new one. Similarly, when we PUT, we're telling the server to update the resource if it's there or create a new one otherwise.

* **Idempotent: Yes (repeated requests have the same effect as a single request).** The PUT method is also Indempotent like the GET method. So, if we send the same update twice, the server will process them in a way that result is the same as if we sent it only once. It is like applying the same correction to a document, doing it once or more does not matter.

* **Example: Updating the content of an article.** A good example for this method type is updating the content of an article or blog. Whenever we write a blog online, and we make a change to it, that change is reflected only once, even if we make that change for more than 1 times. This ensures the correctness of our blog and prevents the content to be harmed by some types of human errors such as multiple similar type changer/alters.

## **04. DELETE:**



image.png

* ***Purpose: Delete a resource identified by the URL.*** The DELETE method is used to request the removal of a resource identified by the URL. In the context of web development, this often means deleting a record or data associated with a specific URL.

* ***Idempotent: Yes (repeated requests have the same effect as a single request).*** Which means, in case of DELETE, if we delete a resource and then try to delete it again, the result will be same, i.e., the resource will be deleted. Thus, DELETE is indempotent.

* ***Example: Deleting a user account.***
Consider a scenario, where we have a web application, and users can have accounts on that application. Now, when a user decides to close their account, the DELETE method can be used to send a request to the server to delete all data associated with that user, effectively closing the account.
