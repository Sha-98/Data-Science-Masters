# API - All about HTTP Methods

### **What is an HTTP method?**
HTTP (Hypertext transfer Protocol) methods, or also HTTP verbs as said, define the actions that can be performed on resources identified by a URL.

Each HTTP method has a specific purpose and is used in different scenarios. This article talks about differen HTTP methods and their purposes in API.

image.png

**Some commonly used HTTP methods are listed and discusses below**

**01. GET:**

* **Purpose: "Retrieve data from the server."**
Which means that when we use the GET method, we are like someone going to the library and asking the librarian for specific book. The librarian gives us the book and we leave with it.

* **Idempotent: Yes (repeated requests have the same effect as a single request).** Now, here the term, 'indempotent' is like saying that if we go back to the librarian and ask for the same book, we'll get the same information. It is like asking for the latest news online. If we refresh the page, we'll get the latest news again. The action of refreshing is 'indempotent'.

* **Example: Retrieving the content of a web page.** As an example, in practical terms, when we type a web address into our browser and hit 'Enter', we're actually sending a GET request to the server hosting that website. The server then responds by sending us the web page content - all the we see.  



**02. POST:**

* **Purpose: Submit data to be processed to a specified resource.**

* **Idempotent:** No (repeated requests may have different effects).
* **Example:** Submitting a form with user data.



**03. PUT:**

* **Purpose:** Update a resource or create a new resource if it does not exist.
* **Idempotent:** Yes (repeated requests have the same effect as a single request).
* **Example:** Updating the content of an article.



**04. DELETE:**

* **Purpose:** Delete a resource identified by the URL.
* **Idempotent:** Yes (repeated requests have the same effect as a single request).
* **Example:** Deleting a user account.



**05. PATCH:**

* **Purpose:** Apply partial modifications to a resource.
* **Idempotent:** No.
* **Example:** Modifying only the email address of a user.

**06. HEAD:**

* **Purpose:** Retrieve the headers for a resource without the body.
* **Idempotent:** Yes.
* **Example:** Checking if a resource has been modified without downloading the entire content.


**07. OPTIONS:**

* **Purpose:** Describe the communication options for the target resource.
* **Idempotent:** Yes.
* **Example:** Checking which HTTP methods are supported on a resource.



**08. TRACE:**

* **Purpose:** Echo the received request, so that a client can see what (if any) changes or additions have been made by intermediate servers.
* **Idempotent:** Yes.
* **Example:** Debugging and diagnostics.



**09. CONNECT:**

* **Purpose:** Establish a tunnel to the server identified by a given URL.
* **Idempotent:** No.
* **Example:** Used with HTTPS proxy servers.



***Understanding these HTTP methods is crucial for designing RESTful APIs and web applications. Each method plays a specific role in the client-server interaction, and their proper use ensures a consistent and efficient web communication.***
