In Pandas, a DataFrame is a two-dimensional, tabular data structure that consists of rows and columns. It is similar to a spreadsheet or SQL table, where each column can be of a different data type. A DataFrame is a primary Pandas data structure for handling and analyzing structured data.

On the other hand, a Series is a one-dimensional labeled array and is one of the fundamental Pandas data structures. It can be thought of as a single column of a DataFrame. A DataFrame is essentially a collection of Series objects.

***An example for this is given below***

```python
# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}

df = pd.DataFrame(data)

# Creating a Series
ages = pd.Series([25, 30, 22], name='Age')
```

## **Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?**


Pandas provides a wide range of functions to manipulate data in a DataFrame. Here are some common functions:

**01. head() and tail():**

* head(n) returns the first n rows of the DataFrame.
* tail(n) returns the last n rows of the DataFrame.

```python
# Load Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Display the first few rows of the dataset
print("Head of the DataFrame:")
titanic_df.head()
```

```python
titanic_df.tail()
```

**02. info():**

Provides a concise summary of the DataFrame, including data types and missing values.

```python
titanic_df.info()
```

**3. describe():**

Generates descriptive statistics of the DataFrame, such as mean, median, and quartiles.

```python
titanic_df.describe()
```

**04. set_index():**

Sets the DataFrame index using one or more existing columns.

```python
titanic_df.set_index('PassengerId', inplace=True)
titanic_df.head()
```

**05. reset_index():**

Resets the DataFrame index to default integer index.

```python
titanic_df.reset_index(drop=True, inplace=True)
titanic_df.head()
```

**06. Filtering the data on a condition**

Filtering the data on a condition on specific column, and storing this new data into a new variable.

```python
female_passengers = titanic_df.loc[titanic_df['Sex'] == 'female']
female_passengers.head()
```

**07. mean():**

Printss the mean value for each numeric column in your DataFrame

```python
age_by_class = titanic_df.groupby('Pclass')['Age'].mean()

print("\nMean age by class:")
age_by_class
```

**08. apply():**

Applies a function along the axis of the DataFrame.

```python
titanic_df['Fare_Double'] = titanic_df['Fare'].apply(lambda x: x * 2)

# Display the DataFrame with the new column
print("\nDataFrame with 'Fare_Double' column:")
titanic_df.head()
```

## **Q6. Which of the following is mutable in nature Series, DataFrame, Panel?**


In Pandas, among Series, DataFrame, and Panel, only DataFrame is mutable in nature. A mutable object can be modified after it is created.

* **Series:** Series Objects are value-mutable but size-immutable objects.

* **DataFrame:** You can modify the contents of a DataFrame, add or remove columns, and perform various operations to change its structure and values.

* **Panel:** Deprecated - The Panel data structure has been deprecated in recent versions of Pandas, and it's recommended to use multi-index DataFrames instead. If you're working with Panel-like data, you would typically use a multi-index DataFrame, which is mutable.

## **Q7. Create a DataFrame using multiple Series. Explain with an example.**

can create a Pandas DataFrame using multiple Series by combining them into a dictionary where each Series becomes a column in the DataFrame. Here's an example:

```python
import pandas as pd

# Creating multiple Series
s1 = pd.Series([1, 2, 3], name='Column1')
s2 = pd.Series(['A', 'B', 'C'], name='Column2')
s3 = pd.Series([10.5, 20.5, 30.5], name='Column3')

# Combining Series into a dictionary
data = {'Column1': s1, 'Column2': s2, 'Column3': s3}

# Creating a DataFrame from the dictionary
df = pd.DataFrame(data)

# Displaying the DataFrame
print("DataFrame created from multiple Series:")
df
```

In this example:

1. We create three Series (s1, s2, and s3) with different data types and names.
2. We combine these Series into a dictionary (data) where the keys are the column names, and the values are the Series.
3. We create a DataFrame (df) from the dictionary, and each Series becomes a column in the DataFrame.
4. Finally, we display the resulting DataFrame.
