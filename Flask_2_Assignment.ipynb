## **Q1. Explain GET and POST methods.**


### **GET**

* **Purpose: "Retrieve data from the server."**
Which means that when we use the GET method, we are like someone going to the library and asking the librarian for specific book. The librarian gives us the book and we leave with it.

* **Idempotent: Yes (repeated requests have the same effect as a single request).** Now, here the term, 'indempotent' is like saying that if we go back to the librarian and ask for the same book, we'll get the same information. It is like asking for the latest news online. If we refresh the page, we'll get the latest news again. The action of refreshing is 'indempotent'.

* **Example: Retrieving the content of a web page.** As an example, in practical terms, when we type a web address into our browser and hit 'Enter', we're actually sending a GET request to the server hosting that website. The server then responds by sending us the web page content - all the we see.  


### **POST**

* **Purpose: Submit data to be processed to a specified resource.** Let's think of POST like a sending package. When we POST something, we're giving new information to the server and asking it to do something with that information. It is like sending a package to a friend; we're delivering new content, the package to our friend's loaction, the server for us and expecting some specific action to be taken.

* **Idempotent: No (repeated requests may have different effects).** Unline GET, POST request or method is not indempotent. Which means, in POST, if we send same request to the server twice, the server might react differently each time, as it might create a new user account for new request, no matter if it is the same request.

* **Example: Submitting a form with user data.** A form we submit on goolge is a good example of the POST method, as here we are making the POST request by sending the data to the sever and asking it to do something with it (it can be any task, as simple as storing also). Everytime we submit the same google form, even if we are adding same info, google considers it as a new user and saves it that way only. This is the POST request for us.


## **Q2. Why is request used in Flask?**


In Flask, the **request** object is used to access incoming request data when handling **HTTP requests**. It allows you to retrieve information sent by the client in the form of parameters, form data, JSON payloads, headers, cookies, and more. The request object is part of the Flask framework and provides a convenient way to interact with the data sent by a client during an HTTP request.

Some common use cases for the 'request' object in flask are:
1. Accessing form data
2. Handling URL Parameters
3. Accessing JSON Data
4. Handling Cookies


## **Q3. Why is redirect() used in Flask?**


In Flask, the **redirect()** function is used to perform an HTTP redirect. It allows you to redirect the client's browser to a different URL, either within your application or to an external URL.
The primary use cases for redirect() include:

1. Page Redirection
2. External URL Redirection
3. Conditional Redirection
4. Redirecting to the Same Endpoint

The redirect() function is a helpful tool for managing the flow of a web application by directing users to different pages based on specific conditions or user actions. It enhances the user experience and allows developers to create dynamic and interactive web applications.


## **Q4. What are templates in Flask? Why is the render_template() function used?**


In Flask, templates are files containing HTML and placeholders for dynamic content that are used to generate dynamic web pages. Templates allow you to separate the structure of your HTML from the Python code in your application, promoting cleaner code and easier maintenance. Flask uses the Jinja2 templating engine by default.

The render_template() function in Flask is used to render templates by combining them with the data provided by your application. This function takes the name of the template file as its first argument and any additional keyword arguments representing variables to be passed to the template.

Here's a brief overview of why templates and the render_template() function are essential in Flask:

**1. Separation of Concerns:**

Templates help separate the HTML markup from the Python code in your application, following the principle of separation of concerns. This makes your code more maintainable and easier to understand.

**2. Dynamic Content:**

Templates allow you to embed dynamic content within HTML pages. By using placeholders or template variables, you can inject data from your Flask application into the HTML, making your web pages dynamic.

**3. Reuse and Consistency:**

Templates can be reused across multiple routes or pages, promoting consistency in the visual presentation of your application. This is particularly useful for creating a consistent layout or navigation structure.

**4. Template Inheritance:**

Flask templates support inheritance, enabling you to create a base template with common elements (header, footer, etc.) and extend or override specific sections in child templates. This enhances code organization and reduces redundancy.

**5. Simplified HTML Generation:**

Using the render_template() function simplifies the process of generating HTML. Instead of manually concatenating strings or using other approaches, Flask integrates seamlessly with Jinja2, making it easy to embed Python logic within your HTML.
