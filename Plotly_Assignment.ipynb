## **Q1. How can you create a Bokeh plot using Python code?**


Creating a Bokeh plot involves importing the necessary libraries and using the Bokeh functions to define and customize the plot. Here's a simple example using Python code:

```python
# Importing necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title='Bokeh Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line glyph to the figure
p.line(x, y, line_width=2)

# Show the plot in a Jupyter notebook (optional)
output_notebook()
show(p)
```

## **Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.**


Glyphs are visual markers that represent data points on a plot. They can be used to display various types of data, such as points, lines, circles, squares, bars, and more.

Glyphs are added to a Bokeh plot using the figure object, and their appearance and properties can be customized.

Bokeh offers a variety of glyphs and customization options, allowing you to create rich and interactive visualizations based on your specific needs.

Example:-

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title='Bokeh Glyph Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a circle glyph to the figure
p.circle(x, y, size=10, color='blue', alpha=0.8, legend_label='Circle Glyph')

# Add a line glyph to the figure
p.line(x, y, line_width=2, line_color='red', legend_label='Line Glyph')

# Customize the plot further (optional)
p.legend.location = 'top_left'

# Show the plot in a Jupyter notebook (optional)
output_notebook()
show(p)
```

## **Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?**


In Bokeh, we can customize the appearance of a plot by modifying various attributes of the figure object.

Here are some common elements we can customize:

**01. Title and Axis Labels:**

Setting the title using the title attribute of the figure object.

Customizing the x-axis and y-axis labels using x_axis_label and y_axis_label, respectively.

```python
p = figure(title='Customized Bokeh Plot', x_axis_label='X-axis Label', y_axis_label='Y-axis Label')
```

**02. Axes Ticks and Grid Lines:**

Customize the appearance of ticks and grid lines using the xaxis and yaxis properties of the figure object.

```python
p.xaxis.major_tick_line_color = 'red'
p.xaxis.major_tick_line_width = 2
p.yaxis.minor_tick_line_color = 'blue'
```

**03. Legend:**

Customize the legend's position, background, and other properties using the legend property.

```python
p.legend.location = 'top_left'
p.legend.background_fill_color = 'lightgray'
```

**04. Plot Background:**

Modify the background color of the plot using the background_fill_color attribute.

```python
p.background_fill_color = 'lightyellow'
```

**05. Grid and Axes Line Colors:**

Change the color of grid lines and axes using the grid property.

```python
p.grid.grid_line_color = 'gray'
p.xaxis.axis_line_color = 'green'
```

**Using all these customizations, an example plot using Bokeh is as follows:**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure with customizations
p = figure(
    title='Customized Bokeh Plot',
    x_axis_label='X-axis Label',
    y_axis_label='Y-axis Label',
    background_fill_color='lightgreen',
)

# Customize grid properties
p.grid.grid_line_color = 'gray'

# Add glyphs or other elements to the plot...
p.line(x, y, line_width=2, legend_label='Line')

# Add a circle glyph to the figure
p.circle(x, y, size=10, color='red', alpha=0.8, legend_label='Circle Glyph')

# Show the plot in a Jupyter notebook (optional)
output_notebook()
show(p)
```

## **Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?**


Bokeh Server is a component of the Bokeh plotting library that allows we to create interactive web applications with real-time updates.

It enables the creation of dynamic and interactive visualizations that respond to user interactions or external data changes.

Bokeh Server facilitates the development of applications where the state of the plot or data can be modified dynamically and pushed to connected clients.

Some key features and steps to use the Bokeh server are:

**01. Interactive Plotting:** Bokeh Server enables the creation of plots that can be updated dynamically based on user interactions, such as selections, sliders, buttons, etc.

**02. Real-Time Updates:** Bokeh Server allows us to update plots in real time, either based on user interactions or external data changes.

**03. Python Callbacks:** We can define Python callbacks that respond to events (like button clicks or slider changes) and update the plot accordingly.

**04. Multiple Users:** Bokeh Server can handle multiple users simultaneously, each interacting with their own instance of the plot.

**05. Deployable:** Bokeh Server applications can be deployed on a server, making them accessible over the web.


An example can be seen as follows:

```python
# Import necessary modules
from bokeh.io import curdoc
from bokeh.models import ColumnDataSource, Slider
from bokeh.models.widgets import Slider
from bokeh.plotting import figure
from bokeh.layouts import column
from bokeh.server.server import Server
from tornado.ioloop import IOLoop

# Create initial data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y})

# Create a figure
plot = figure(title='Bokeh Server Example', plot_height=300, plot_width=400)
line = plot.line('x', 'y', source=source, line_width=2)

# Define a callback function
def callback(attr, old, new):
    # Update the data when the slider is interacted with
    source.data['y'] = [val * slider.value for val in x]

# Create a slider widget
slider = Slider(start=1, end=10, step=1, value=1, title='Multiplier')
slider.on_change('value', callback)

# Arrange the layout
layout = column(slider, plot)

# Add the layout to the current document
curdoc().add_root(layout)

# Run the application
if __name__ == '__main__':
    server = Server({'/': layout}, io_loop=IOLoop(), allow_websocket_origin=["localhost:8000"])
    server.start()
    server.io_loop.start()
```

## **Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(width=600, height=400, x_axis_label='x', y_axis_label='y',
           tools="pan,box_zoom,reset,save")
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

show(p)
```

```python
from bokeh.embed import components
from bokeh.plotting import figure

# Defining Chart Data
language = [
	'Python', 'Java', 'JavaScript', 'C#', 'PHP', 'C/C++',
	'R', 'Objective-C', 'Swift', 'TypeScript', 'Matlab',
	'Kotlin', 'Go', 'Ruby', 'VBA'
]
popularity = [
	31.56, 16.4, 8.38, 6.5, 5.85, 5.8, 4.08, 2.79, 2.35,
	1.92, 1.65, 1.61, 1.44, 1.22, 1.16
]

# Creating Plot Figure
p = figure(
	x_range=language,
	height=400,
	title="Popularity of Programming Languages",
	sizing_mode="stretch_width"
)

# Defining Plot to be a Vertical Bar Plot
p.vbar(x=language, top=popularity, width=0.5)
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Get Chart Components
script, div = components(p)

print(script)
print(div)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Data
popularity = [31.56, 16.4, 8.38, 6.5, 5.85, 5.8, 4.08, 2.79, 2.35, 1.92, 1.65, 1.61, 1.44, 1.22, 1.16]
language = ['Python', 'Java', 'JavaScript', 'C#', 'PHP', 'C/C++', 'R', 'Objective-C', 'Swift', 'TypeScript', 'Matlab', 'Kotlin', 'Go', 'Ruby', 'VBA']

# Create a figure
p = figure(x_range=language, title='Popularity of Programming Languages')

# Add vertical bars
p.vbar(x=language, top=popularity, width=0.9, color="blue")

# Styling
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = max(popularity) + 5

# Display the plot
output_notebook()
show(p)
```
