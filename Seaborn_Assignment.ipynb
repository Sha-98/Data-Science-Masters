## **Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**



**Seaborn** is a powerful data visualization library in Python that works on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn:

**01. Histograms (sns.histplot()):**

Use: Display the distribution of a continuous variable.

Example: Visualizing the distribution of ages in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 25, 35, 40, 45, 45, 50, 55, 60, 25]

# Create a histogram
sns.histplot(ages, bins=10)
plt.show()
```

**02. Scatter Plots (sns.scatterplot()):**

Use: Show the relationship between two continuous variables.

Example: Investigating the correlation between hours studied and exam scores.

```python
hours_studied = [2, 3, 4, 5, 6, 7, 8, 4, 5, 6, 9 , 5 , 4]
exam_scores = [65, 70, 75, 80, 85, 90, 95, 52, 89, 74, 89, 45, 56]

# Create a scatter plot
sns.scatterplot(x=hours_studied, y=exam_scores)
plt.show()
```

**03 Bar Plots (sns.barplot()):**

Use: Display the relationship between a categorical variable and a continuous variable.

Example: Comparing average sales across different product categories.

```python
categories = ['Category A', 'Category B', 'Category C']
average_sales = [500, 700, 600]

# Create a bar plot
sns.barplot(x=categories, y=average_sales)
plt.show()
```

**04. Box Plots (sns.boxplot()):**

Use: Visualize the distribution and spread of a continuous variable across different categories.

Example: Examining the distribution of scores in different subject areas.

```python
import pandas as pd

# Sample data
subjects = ['Math', 'Science', 'English']
scores_math = [80, 85, 90, 95, 100]
scores_science = [75, 80, 85, 90, 95]
scores_english = [70, 75, 80, 85, 90]

# Create a DataFrame
data = {'Math': scores_math, 'Science': scores_science, 'English': scores_english}
df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(data=df)
plt.show()
```

**05. Heatmaps (sns.heatmap()):**

Use: Visualize the correlation matrix of variables in a dataset.

Example: Understanding the relationships between multiple features.

```python
import numpy as np
corr_matrix = np.random.rand(5, 5)

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

## **Que 2: Load the "fmri" dataset using the load_dataset function of seaborn.**

**Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

**Note: timepoint, signal, event, and region are columns in the fmri dataset.**




```python
# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Show the plot
plt.show()
```

## **Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.**

**Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.**

**Note: pclass, age, and fare are columns in the titanic dataset.**



```python
# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for 'age' and 'fare' based on 'pclass'
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
```

## **Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**


```python
# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram for 'price' with 'cut' as hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30, kde=True)

# Set plot title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')

# Show the plot
plt.show()
```

## **Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**


```python
# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot a pair plot with 'species' as hue
sns.pairplot(iris_data, hue='species', markers=['o', 's', 'D'])

# Set plot title
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02)

# Show the plot
plt.show()
```

## **Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_matrix = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot a heatmap
sns.heatmap(flights_matrix, annot=True, fmt="d", cmap="YlGnBu")

# Set plot title
plt.title('Flights Dataset - Heatmap')

# Show the plot
plt.show()
```
