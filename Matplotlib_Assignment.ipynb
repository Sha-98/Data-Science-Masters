## **Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**


***Matplotlib*** is a 2D plotting library for Python that produces publication-quality figures in various formats and interactive environments across platforms. It is often used for data visualization and creating static, animated, and interactive plots in Python.

**Key features of Matplotlib:**

**01. Comprehensive:** Matplotlib provides a wide variety of plots and charts, making it versatile for different data visualization needs.

**02. Customizable:** Users have fine-grained control over aspects like line styles, colors, labels, and more.

**03. Integration:** Matplotlib can be used seamlessly with other libraries, such as NumPy and Pandas, for data manipulation and analysis.

**04. Publication-Quality:** Matplotlib produces high-quality plots suitable for publication and presentation.

***Pyplot***, a sub-module of Matplotlib, provides a simple interface for creating common types of plots. Here are five types of plots that can be created using the Pyplot module:



***01. Line Plot: Visualizes data points connected by straight line segments.***

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

***02. Scatter Plot: Displays individual data points without connecting lines.***

```python
x = [1.2, 0.5, 1, 1.3, 2.5]
y = [2, 2.7, 3.6, 1, 1.0]

plt.scatter(x, y)
plt.show()
```

***03. Bar Plot: Represents data with rectangular bars.***

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 5]

plt.bar(categories, values)
plt.show()
```

***04. Histogram: Visualizes the distribution of a dataset.***

```python
import numpy as np

data = np.random.randn(1000)  # Example random data

plt.hist(data, bins=30)
plt.show()
```

***05. Pie Chart: Represents data in a circular statistical graphic.***

```python
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

## **Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**



```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

***Note: Also add title, xlabel, and ylabel to the plot.***

**Scatter plot** is a type of plot or mathematical diagram that uses Cartesian coordinates to display values for two variables for a set of data.

Each point in the dataset is represented by a dot, and the position of the dot on the horizontal and vertical axis corresponds to the values of the two variables.

```python
plt.scatter(x, y, label='Scatter Plot', color='blue', marker='o')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```

## **Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:**

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The **subplot()** function in Matplotlib is used to create a grid of subplots within a single figure. This allows you to display multiple plots in the same figure, making it easier to compare and analyze different datasets or aspects of your data.

The **subplot()** function takes three arguments: the number of rows, the number of columns, and the index of the subplot you want to create.

```python
# Create a single subplot
plt.subplot(1, 1, 1)

# Data for Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x1, y1, label='Line 1')

# Data for Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x2, y2, label='Line 2')

# Data for Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x3, y3, label='Line 3')

# Data for Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x4, y4, label='Line 4')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines on a Single Graph')
plt.legend()

# Display the plot
plt.show()
```

## **Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**


A **bar plot** is a visual representation of data using rectangular bars or columns. Each bar's length or height corresponds to the value it represents, making it easy to compare different categories.

Bar plots are commonly used to display and compare the values of different groups or categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Profit Comparison by Company (Bar Plot)')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Profit Comparison by Company (Horizontal Bar Plot)')
plt.show()
```

## **Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A **box plot**, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset.

It displays the summary statistics of a dataset, including the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum. The "box" represents the interquartile range (IQR), and the "whiskers" extend from the box to indicate the range of the data.

**Box plots** are useful for identifying the central tendency, spread, and skewness of a dataset. They are particularly helpful in comparing the distribution of multiple datasets.

```python
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Two Boxes')
plt.show()
```
