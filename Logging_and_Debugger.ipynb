## Logging in Python

image.png

Logging in computer science world, is a crucial aspect of software development. It involves capturing and recording information about a program's runtime behavior in a log file. In this case, python provides a build-in logging module that makes it easy to implement logging in our applications to check, know, and record what is going on in the compiler at runtime. The 'logging' module allows us to record information about events, and errors encountered in the code at runtime, which can be very useful for debugging, monitoring, and auditing purposes.

### Key componenets of Python Logging

#### Logger
A logger is an object in python which represents a named channel for log messages. We create and configure logger objects to determine where log messages should be recorded and at what severity level they should be captured and added in the log file.

#### Handler
A handler is responsible for processing log messages and determining whether they should be the output. Oython provides us various build-in handlers, such as "StreamHandler" which gives us console output, and "FileHandler" which gives us file output. Apart from these, we can also create custom handlers.

#### Formatter
Formatter specifies the format of log messages, including the information like timestamp, log level, and the actual message of the log. We can always customize the log message format to suit our needs at the momemt in project.

Talking about the logging levels in python, we get a series of logging levels in the logging module in python.

### Logging Levels

Python's 'logging' module defines several logging levels, which allow us to categorise log messages by their severity. The commonly used logging levels, in the ascending order of their avaialability in python are explained below.

1. 'DEBUG' : Detailed information for debugging process
2. 'INFO' : General information about the program's operation.
3. 'WARNING' : Indicates that something unwanted or unexpected or potentially problematic occured.
4. 'ERROR' : Errors that prevented certain operations from completing.
5. 'CRITICAL' : Critical errors that may lead to program termination

The syntax and examples of using or generating logging document in python using the 'logging' module can be seen in the cells below.



```python
import logging
```

```python
# configuring the logging system
logging.basicConfig(filename = 'projLog.log', level=logging.DEBUG)

# creating a logger
logger = logging.getLogger('my_logger')

# defining custom log messages
logger.debug('This is a debug message')
logger.info('This is an info message')
logger.warning('This is a warning message')
logger.error('This is an error message')
logger.critical('This is a critical message')
```

One thing is to be noted here is that the logging levels available for you depends upon the level mentioned at the time of configuring the logging system.

In the code cell above, we defined the level as 'DEBUG' and we were able to access all the levels through the logger.

But this would not be the case if the level was set to be 'WARNING', or 'ERROR'

The logger can only access the level equal to and below the one mentioned at the time of configuring, but not above them. So, if we have mentioned the level as 'WARNING', our logger will not be able to access the 'DEBUG' and 'INFO' level, but all the others. It would not throw an error, but the logger would also not capture the data for unaccessible levele and add them in the log file.

### Advantages of Logging in Real Projects

Logging is a crucial practice in real-world software development for the following reasons.

1. Debugging:
Logging helps developers identify and diagnose issues in the code during the development and testing process of applications. Detailed log messages can provide insights into the program's behavior and state.

2. Monitoring:
In production environment, logging allows us to monitor the health and performance of applications. Log messages can be collected and analysed to detect anomalies or errors in real-time.

3. Auditing and Compliance:
Logging is essential for auditing and ensuring compliance with regulatory requirements. It also provides a record of actions taken within software, which can be useful for security legal purposes.

4. Troubleshooting:
When issues arise in deployed systems, logs provide valuable information for troubleshooting and identifying the root cause of problems.

5. Error Reporting:
In web applications and services, logs can be used to capture and report errors to developers or support teams, making it easier to address issues reported by users.

6. Historical Data:
Log files serve as a historical record of a program's behavior. They can be archived and analyzed to gain insights into long-term trends or patterns.

7. Flexibility:
Python's logging module is highly configurable, allowing you to customize log message formats, output destinations, and logging levels to suit your project's requirements.


In real project teams and industries, effective logging practices help ensure the reliability, maintainability, and security of software systems. It provides transparency into the behavior of complex applications and facilitates collaborative debugging and issue resolution.
